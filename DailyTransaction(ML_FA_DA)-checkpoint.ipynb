```python
sns.countplot(data = df, x = "Category", order =
df["Category"].value_counts().iloc[:5].index)
```

```python
plt.figure(figsize=(10, 6))

top5 = df["Category"].value_counts().iloc[:5].index

sns.countplot(
    data=df,
    x="Category",
    order=top5,
    hue="Category",     # map each bar to a color
    palette="tab10",    # or Set3, husl, Paired
    legend=False        # hide legend
)

plt.show()
```

```python
df["Subcategory"].unique()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,x="Subcategory",order=
df["Subcategory"].value_counts().iloc[:10].index)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

top10 = df["Subcategory"].value_counts().iloc[:10].index

sns.countplot(
    data=df,
    x="Subcategory",
    order=top10,
    hue="Subcategory",   # assign hue
    palette="tab10",     # or Set3, husl, Paired
    legend=False         # remove legend
)

plt.xticks(rotation=90)
plt.show()
```
